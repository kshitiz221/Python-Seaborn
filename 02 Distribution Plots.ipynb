```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%config Completer.use_jedi = False
```

## Distribution Plots

* Show single continous features
* Visualize average and deviation, other such values
    * Rug
    * Histogram
    * Kernel Density Estimation (KDE)
        * Method of estimating a probability density function of a random variable, for finite data

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.set(style='whitegrid') 
sns.displot(data=df, x='salary',bins=25) # more generalized
```

```python
sns.set(style='whitegrid')
sns.displot(data=df, x='salary', bins=20, color='magenta', edgecolor='green', linewidth=1.5, kde=True)
```

```python
sns.histplot(data=df, x='salary', bins=20, color='magenta', edgecolor='green', linewidth=1.5, kde=True)
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=["age"])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age', height=0.5)
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages,x='age',kde=True, bins=20)
```

```python
sns.kdeplot(data=sample_ages,x='age')
```

##### Cut Off KDE

We could cut off the KDE if we know our data has hard limits (no one can be a negative age and no one in the population can be older than 100 for some reason)

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100])
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.2)
```

```python
sns.kdeplot(data=sample_ages,x='age',bw_adjust=0.5, shade=True, color='purple')
```
