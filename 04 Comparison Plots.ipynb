```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%config Completer.use_jedi = False
```

## Comparison Plots

* 2-dimensional version
* Two main plots:
    * jointplot()
    * pairplot()

```python
dataset = pd.read_csv("StudentsPerformance.csv")
```

```python
dataset.head()
```

### Joint Plot

* map histograms to each feature of a scatter plot
* adjust scatterplot to be a hex or 2D KDE plot

```python
sns.jointplot(data=dataset, x='math score', y='reading score')
```

```python
sns.jointplot(data=dataset, x='math score', y='reading score', kind='scatter', color='purple', hue='gender')
# kind - hist, kde, scatter, hex, reg, resid
# hex does not support HUE
```

```python
sns.jointplot(data=dataset, x='math score', y='reading score', kind='kde', color='purple', shade=True)
# kind - hist, kde, scatter, hex, reg, resid
```

### Pair Plot

* compare all numerical columns in a DataFrame
* automatically creates histogram for each column
* automatically creates scatterplots between possible columns
* for large datasets, first filter out columns for pairplot
* CPU & RAM intensive

```python
sns.pairplot(dataset)
```

```python
sns.pairplot(dataset, hue='gender')
#sns.pairplot(dataset, hue='gender', diag_kind='hist')
```

```python
sns.pairplot(dataset, hue='gender', corner=True, palette='viridis')
# removes duplicate plots
```
