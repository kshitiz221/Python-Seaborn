```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%config Completer.use_jedi = False
```

## Grids

* Uses matplotlib() subplots() to automatically create a grid based off a categorical column
* Simply pass in the name of the column and seaborn will automatically map the subplot grids

```python
dataset = pd.read_csv("StudentsPerformance.csv")
```

```python
dataset.head()
```

### catplot()

```python
sns.catplot(data=dataset, x='gender', y='math score', kind='violin', row='lunch')
# Kind: “point”, “bar”, “strip”, “swarm”, “box”, “violin”, or “boxen”
```

```python
sns.catplot(data=dataset, x='gender', y='math score', kind='violin', col='lunch')
```

```python
sns.catplot(data=dataset, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

### Pair Grid
* Grid that pairplot is built on top of, allows for heavy customization of the pairplot

```python
grid = sns.PairGrid(dataset, hue='gender', palette='Set2')
grid = grid.map_upper(sns.scatterplot)
grid = grid.map_lower(sns.kdeplot)
grid = grid.map_diag(sns.kdeplot, lw=2)
grid = grid.add_legend()
```

### Facet Grid

```python
sns.FacetGrid(data=dataset, col='gender', row='lunch')
```

```python
g = sns.FacetGrid(data=dataset, col='gender', row='lunch')
g = g.map(plt.scatter, "math score", "reading score", edgecolor="w")
g.add_legend()
plt.subplots_adjust(hspace=0.4, wspace=0.5)
```
