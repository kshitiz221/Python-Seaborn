```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%config Completer.use_jedi = False
```

## Matrix Plots

* Visual equivalent of displaying a pivot table
* Displays all the data passed in, visualizing all the numeric values in a DataFrame
* Types: 
    * heatmap()
    * clustermap()

```python
dataset = pd.read_csv("country_table.csv")
```

```python
dataset.head()
```

```python
dataset = dataset.set_index('Countries')
dataset              
```

### heatmap()
* visually displays the distribution of cell values with a color mapping

```python
plt.figure(dpi=100)
sns.heatmap(dataset)
# rates are percentage, whereas life expectancy in in years
# visually not informative
```

```python
plt.figure(dpi=100)
sns.heatmap(dataset.drop('Life expectancy', axis=1), linewidth=0.5, annot=True)
```

```python
plt.figure(dpi=100)
sns.heatmap(dataset.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='plasma')
```

### clustermap()
* same visual as heatmap
* first conducts hierarchical clustering to recognize data into groups

```python
plt.figure(dpi=100)
sns.clustermap(dataset.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='plasma')
```

```python
plt.figure(dpi=100)
sns.clustermap(dataset.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='plasma', col_cluster=False)
```
