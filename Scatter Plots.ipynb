```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%config Completer.use_jedi = False
```

## Scatter Plots

* Show relationship between two continous features
* Continous featuresa are numeric variables that can take any number of values between any two values
* Example
    * Age
    * Height
    * Salary
    * Temperature
    * Prices


```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.heatmap(df.corr())
```

```python
plt.figure(figsize=(10,8), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
plt.figure(figsize=(12,8), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education')
```

```python
plt.figure(figsize=(12,8), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='Dark2') # shows a gradient
```

```python
plt.figure(figsize=(12,8), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, size='salary')
# low salary, small points
# High salary, big points
```

```python
plt.figure(figsize=(12,8), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, alpha=0.2, s=200)
```

```python
plt.figure(figsize=(12,8), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, style='level of education', hue='level of education')
plt.savefig('Scatterplot.jpeg')
```
